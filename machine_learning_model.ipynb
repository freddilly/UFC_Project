## Data Cleaning

```python
#Look at database makeup
fight_data_pre_df.isna().sum()
```

```python
#Drop certain columns
fight_data_prerun = fight_data_pre_df.copy() 
fight_data_prerun = fight_data_prerun.drop("r_record",axis=1)
fight_data_prerun = fight_data_prerun.drop("r_takedown_diff",axis=1)
fight_data_prerun = fight_data_prerun.drop("b_record",axis=1)
fight_data_prerun = fight_data_prerun.drop("b_takedown_diff",axis=1)
fight_data_prerun = fight_data_prerun.drop("r_fighter",axis=1)
fight_data_prerun = fight_data_prerun.drop("r_stance",axis=1)
fight_data_prerun = fight_data_prerun.drop("b_stance",axis=1)
fight_data_prerun = fight_data_prerun.drop("b_fighter",axis=1)
fight_data_prerun = fight_data_prerun.drop("title_bout",axis=1)
```

```python
fight_data_to_run = fight_data_prerun.dropna()
```

```python
fight_data_to_run
```

## Machine Learning Model

```python
#import more dependencies
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

```python
#Define the X variables
X_pre = fight_data_to_run
X = X_pre.drop("winner",axis=1)
```

```python
y = fight_data_to_run["winner"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Train the data
classifier.fit(X_train_scaled,y_train)
```

```python
# Predict outcomes for test data set
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
acc_score
```
